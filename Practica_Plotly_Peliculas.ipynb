**Objetivo: practicar algunos graficos y analizar usando plotly, con el dataset 'netflix_titles' extraido de Kaggle.**

```python
import pandas as pd
import plotly.express as px
```

```python
peliculas = pd.read_csv('netflix_titles.csv')
peliculas.head()
```

```python
peliculas.isnull().sum()  #Me fijo si hay valores nulos
```

```python
peliculas.dropna(subset = ['date_added','rating', 'duration', 'country'], inplace=True) #elimino algunos valores nulos de esas columnas
```

```python
peliculas['director'].fillna('Desconocido', inplace = True)
peliculas['cast'].fillna('Desconocido', inplace = True)
  #Decido remplazar los valores nulos de estas columnas y los relleno con 'Desconocido'
```

```python
peliculas.isnull().sum()    #Reviso si hay valores nulos
```

```python
peliculas = peliculas.rename(columns= {
    	'cast': 'reparto',
      'country': 'pais',
     'date_added': 'fecha_agregada',
     'release_year': 'fecha_estreno',
     'duration' : 'duracion',
     'listed_in': 'categoria'

})  #Cambio algunos nombres a espa√±ol
```

```python
conteo_peliculas = peliculas['type'].value_counts().reset_index()    #Creo un nuevo Df para contar los tipos de peliculas y shows, luego grafico
```

Grafico 1. Pie/torta

```python
fig = px.pie(conteo_peliculas, names='type', values='count', title= 'Cantidad de peliculas vs series', color='type', template= 'plotly_dark')
fig.show()  #observamos en el grafico pie, que hay mayor cantidad de peliculas que tvShows
```

```python
conteo_pais = peliculas['pais'].value_counts().reset_index()
conteo_pais    #Hago un nuevo Df para contar la cantidad de paises con mas peliculas
```

Grafico 2. Barras

```python
fig = px.bar(conteo_pais.head(10), x='pais', y='count', title='Top 10 paises con mas contenido', color='pais', template='plotly_dark' )
fig.show()
```

```python
clasificacion = {
    "rating": {
        "TV-MA": "Adultos",
        "TV-14": "14+",
        "TV-Y7": "7+",
        "PG-13": "13+",
        "R": "17+",
        "G": "Todos",
        "TV-Y": "Peques",
        "PG": "Gu√≠a padres",
        "TV-PG": "Gu√≠a padres",
        "TV-G": "Todos",
        "NR": "Sin clasificar",
        "TV-Y7-FV": "7+ Violencia",
        "NC-17": "Solo adultos",
        "UR": "Sin clasificar"
    }
}
peliculas.replace(clasificacion, inplace=True)   #Cambio los datos de la columna rating para mas claridad
```

```python
peliculas.rename(columns = {'rating': 'clasificacion'}, inplace=True) #cambio el nombre de la columna rating a clasificacion para mayor claridad
```

Grafico 3. Barras horizontales.

```python
agrupar_clasificacion_tipo = peliculas.groupby(['type', 'clasificacion']).size().reset_index(name='count')

fig = px.bar(
    agrupar_clasificacion_tipo,
    x='count',
    y='clasificacion',
    color='type',
    barmode='group',
    orientation='h',
    title='Cantidad de pel√≠culas y series por clasificaci√≥n'
)

fig.show()
#Aqu√≠ observamos que en ‚ÄòPeques‚Äô hay m√°s TV shows que pel√≠culas, a diferencia de otras clasificaciones donde predominan las pel√≠culas.
```

Grafico 4, lineal

```python
conteo_cruzado = peliculas.groupby(['fecha_estreno', 'type']).size().reset_index(name='count')

fig = px.line(
    conteo_cruzado,
    x='fecha_estreno',
    y='count',
    color='type',
    title='Cantidad de pel√≠culas y series por a√±o',
    template='plotly_dark'
)

fig.show()
```
